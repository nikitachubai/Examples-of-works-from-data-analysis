## 6. Split the data from "x" and "y" into a training and test sample. To do this, use the "train_test_split" function. It takes the two previously specified variables x and y as arguments, and here we also define the size of the test sample (usually no more than 20%). To do this, we assign our value in the format "0.20" (which means 20%) to the named argument "test_size". ### 7. After preparing the data set, design the neural network model. To do this, create a class named "Perceptron". For this class, define the "build" method with two parameters:
1) dim (input size of the training sample),
2) num_classes - the number of classes (in our case, there are 3) Next, in the body of this method, create a variable model, where we save an instance of the Sequential class from the Keras-Tensorflow library. Then, using the "add" method, we build the fully connected layers (Dense) of the neural network one by one. The first layer (input layer) will have 128 nodes (neurons) and the ReLU activation function. This layer also accepts the above-mentioned dim argument as a parameter. The second (hidden) layer has 32 neurons and the ReLU activation function. The third layer that completes the network has 2 neurons (according to the number of classes) and the Softmax activation function.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Activation, Dropout
from tensorflow.keras.optimizers import Adam
```

```python

class Perceptron:
    @staticmethod
    def build (dim, num_classes):
        model = Sequential()
        model.add(Dense(128, input_dim=dim, activation='relu'))
        model.add(Dense(32, activation='relu'))
        model.add(Dense(num_classes, activation='softmax'))
        
        return model
```

## 8. Now create an instance of the Perceptron class, storing it in a variable named model (you can use any variable name, but it is recommended to specify names that will most clearly describe their content). Don't forget to define the parameters of the Perceptron class: num_classes should match the number of classes in the dataset, and for dim we pass the input sample size x_train.shape[1], which is 753.

```python
print(X_train.shape)
```

```python
model = Perceptron.build(X_train.shape[1],2)
```

### Output
Created an instance of the Perceptron class (our model)

## 9. Compile the newly created neural network model. To do this, use the “compile” method (Keras-Tensorflow). As its first parameter, specify which error function to use (for example, “sparse_categorical_crossentropy”, which calculates the crossentropy loss between labels and predictions). As the second parameter, specify the optimizer (for example, “adam” or stochastic gradient descent). For information on loss functions and optimizers, see the Keras, Tensorflow documentation. And as the third parameter, specify the target metric – accuracy.

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(0.00001), metrics=['accuracy'])
print("Model compiled")
```

### Output:
Model compiled.

## 10. Now, after completing all the preparatory operations, you can begin the process of training the ANN model. To do this, apply the fit method (Keras, Tensorflow) to the model variable and specify the applied samples (x_train, y_train), batch size, number of iterations (epochs), size of the validation sample "validation_split" (which is created automatically by this fit method) as parameters, and also define the verbose parameter (by setting it to 0, 1 or 2, we choose how we want to "see" the training progress for each epoch).

```python
def print_evaluate(true, predicted):  
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
    print('__________________________________')
    
def evaluate(true, predicted):
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    return mae, mse, rmse, r2_square
```

```python
test_pred = model.predict(X_test)
train_pred = model.predict(X_train)


df_temp1 = pd.DataFrame(test_pred, columns=['1', '2'])
df_temp2 = pd.DataFrame(train_pred, columns=['1', '2'])

print('Test set evaluation:\n_____________________________________')
print_evaluate(y_test, df_temp1['2'])

print('Train set evaluation:\n_____________________________________')
print_evaluate(y_train, df_temp2['2'])

results_df = pd.DataFrame(data=[["Artficial Neural Network", *evaluate(y_test, df_temp1['2']), 0]], 
                            columns=['Model', 'MAE', 'MSE', 'RMSE', 'R2 Square', 'Cross Validation'])
results_df.head()
```
