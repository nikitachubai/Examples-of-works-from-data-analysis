# Import bonus data and calculate the total amount of bonuses for each user

```python
# Import bonus data and calculate the total amount of bonuses for each user
df_bonus = pd.read_csv(r"/////.csv", sep=';')
filtered_data_plus_all = df_bonus.loc[df_bonus['BalanceChangeAmount'] >= 0, :]
filtered_data_minus_all = df_bonus.loc[df_bonus['BalanceChangeAmount'] < 0, :]
pivot_table_plus_all = pd.pivot_table(filtered_data_plus_all, values='BalanceChangeAmount', index='BalanceChangeCustomerIdsMindboxId', aggfunc='sum').reset_index()
pivot_table_minus_all = pd.pivot_table(filtered_data_minus_all, values='BalanceChangeAmount', index='BalanceChangeCustomerIdsMindboxId', aggfunc='sum').reset_index()
```
